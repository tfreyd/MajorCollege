```python
raw_data_count=recent_grads.count()
```

```python
#we remove the data with missing data, important for matplotplib
recent_grads=recent_grads.dropna()
```

```python
cleaned_data_count=recent_grads.count()
```

```python
raw_data_count - cleaned_data_count
```

```python
recent_grads.plot(x='Sample_size', y='Employed', kind='scatter')
```

```python
recent_grads.plot(x='Sample_size', y='Employed', kind='scatter', title='Employed vs. Sample_size', figsize=(5,10))
```

```python
recent_grads.plot('Sample_size','Median',kind='scatter', title='sample size vs median')
```

* there are no correlation betwwen the sample size and the average salary

```python
recent_grads.plot('Sample_size','Unemployment_rate',kind='scatter', title='sample size vs unemployment ')
```

* no relation btw sample size and unployement 

```python
recent_grads.plot('Full_time','Median',kind='scatter', title='# full time students vs salary median')
```

```python
recent_grads.plot('ShareWomen','Median',kind='scatter', title='% of the women in major vs salary median')
```

* it looks that there is small inverse relation between the median salary and the number of women is the sample size.

```python
recent_grads.plot('Women','Median',kind='scatter')
```

* no relation bet women and median
